```python
import os
import subprocess

def convert_ipynb_to_pdf_webpdf(root_dir='.', recursive=True):
    print(f"\nüêç Converting notebooks in: {os.path.abspath(root_dir)}")
    print("    Conversion: .ipynb ‚Üí PDF (Chromium rendering)\n")

    for dirpath, _, filenames in os.walk(root_dir):
        for file in filenames:
            if file.endswith('.ipynb'):
                ipynb_path = os.path.join(dirpath, file)
                pdf_path = os.path.splitext(ipynb_path)[0] + '.pdf'

                if os.path.exists(pdf_path):
                    print(f"    Skipping (already exists): {pdf_path}")
                    continue

                print(f"    Converting: {ipynb_path}")

                try:
                    # Single-step conversion using Chromium
                    subprocess.run([
                        'jupyter', 'nbconvert',
                        '--to', 'webpdf',
                        '--allow-chromium-download',  # Auto-download if needed
                        ipynb_path
                    ], check=True)
                    
                    print(f"    ‚úì Successfully converted: {pdf_path}\n")

                except Exception as e:
                    print(f"    ‚ùå Error converting {ipynb_path}: {e}")
        
        if not recursive:
            break

# Run the conversion
convert_ipynb_to_pdf_webpdf(root_dir='.', recursive=True)
```
